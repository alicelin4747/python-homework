```python
# import Path function from python library and import CSV

from pathlib import Path
import csv

# set path to data file

csvpath = Path('Resource/budget_data.csv')

# Initialize variables
max_pnl_change = 0
min_pnl_change = 0
avg_pnl_change = 0
count_pnl_change = 0
total_pnl_change = 0
prior_pnl = None
current_pnl = 0
line_num = 0
count_pnl = 0
total_pnl = 0

# Open data file in csv format with , as seperator
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter = ',')
    # read the first line and set to variable header
    header = next(csvreader)

# Read each row of data after the header
    for row in csvreader:
        current_pnl = int(row[1])
        current_date = row[0]
        
        # calculate total pnl and pnl count
        count_pnl += 1
        total_pnl += current_pnl
        
        # On the very first row, prior_pnl is None.
        if prior_pnl is not None:
            # calcuate the difference
            change_pnl = (current_pnl - prior_pnl)
            
            # calculate the max and min change
            if change_pnl == 0:
                max_pnl_change = change_pnl
                max_pnl_change_date = current_date
            elif change_pnl > max_pnl_change:
                max_pnl_change = change_pnl
                max_pnl_change_date = current_date
                
            if change_pnl == 0:
                min_pnl_change = change_pnl
                min_pnl_change_date = current_date
            elif change_pnl < min_pnl_change:
                min_pnl_change = change_pnl
                min_pnl_change_date = current_date
            
            # calculate the total change and change count
            count_pnl_change += 1
            total_pnl_change += change_pnl
            
            # calculate the average change
            avg_pnl_change = round(total_pnl_change / count_pnl_change,2)
            
        # On subsequent rows, prior_pnl is the pnl from the previous row
        prior_pnl = current_pnl
        
# Print the analysis
print('Financial Analysis')
print('----------------------------')
print(f'Total Months: {count_pnl}')
print(f'Total: ${total_pnl}')
print(f'Average Change: ${avg_pnl_change}')
print(f'Greatest Increase in Profits: {max_pnl_change_date} (${max_pnl_change})')
print(f'Greatest Decrease in Profits: {min_pnl_change_date} (${min_pnl_change})')

# Set output file path
output_path = Path('output.txt')

# Open the output path as a file object
with open(output_path, 'w') as text_file:
    text_file.write('Financial Analysis\n')
    text_file.write('----------------------------\n')
    text_file.write('Total Months:' + str(count_pnl) + '\n')
    text_file.write('Total: $' + str(total_pnl) + '\n')
    text_file.write('Average Change: $' + str(avg_pnl_change) + '\n')
    text_file.write('Greatest Increase in Profits: ' + str(max_pnl_change_date) + ' ($' + str(max_pnl_change) + ')' + '\n')
    text_file.write('Greatest Decrease in Profits: ' + str(min_pnl_change_date) + ' ($' + str(min_pnl_change) + ')' + '\n')
    
text_file.close()

```
